## Ours

```python
simulation_duration = 1 # seconds

# Neuron parameters
tau_m = 20e-3       # membrane time constant [s]
C_m = 200e-12       # membrane capacitance [F]
v_rest = -70e-3     # resting potential [V]
v_reset = -70e-3    # reset potential after spike [V]
v_thresh = -50e-3   # threshold [V]
refractory = 5e-3   # refractory period [s]

neuron = pb.LIFNeuron(tau_m=tau_m, C_m=C_m, v_rest=v_rest, v_reset=v_reset, v_thresh=v_thresh, refractory=refractory)

# Network parameters
num_neurons = 500
connection_prob = 0.2

# Weights for normal operation test
weight_range_exc = (1.5e-12, 1.5e-12)
weight_range_inh = (-2e-12, -2e-12)

# Weigths for spike saturation test (strong excitatory behavior)
#weight_range_exc = (5e-12, 5e-12)
#weight_range_inh = (0, 0)

delay_range = (0.001, 0.002)
inhibitory_ratio = 0.2
num_inhibitory = int(num_neurons * inhibitory_ratio)

# Generate the Erdős–Rényi random graph with random weights and delays
graph = []
for pre in range(num_neurons):
    for post in range(num_neurons):
        if pre != post:
            if random.random() < connection_prob:
                # Determine if the presynaptic neuron is inhibitory
                if pre >= num_neurons - num_inhibitory:
                    weight = random.uniform(*weight_range_inh)
                else:
                    weight = random.uniform(*weight_range_exc)
                delay = random.uniform(*delay_range)
                graph.append((pre, post, weight, delay))

# Generate spike events for all neurons and store them in a middle storage
# Parameters
spike_rate = 40.0  # spikes per second per neuron
t_start = 100e-6
t_end = 5.0
input_neurons = 10
spike_storage = []
for neuron_id in range(input_neurons):
    t = t_start
    while t < t_end:
        # Inter-spike interval sampled from exponential distribution
        isi = random.expovariate(spike_rate)  # mean interval = 1 / spike_rate
        t += isi
        rounded_t = round(t, 4)  # round to 100 us to be compatible with all simulators
        if rounded_t < t_end:
            magnitude = 30.0*C_m  # spike magnitude
            spike_storage.append((rounded_t, neuron_id, magnitude))
```

```python
# Initialize network and monitors
nn = pb.NeuralNetwork()
nn.set_num_exec_threads(num_threads)
spike_monitor = pb.SpikeMonitor()
nn.set_spike_monitor(spike_monitor)
state_monitor = pb.StateMonitor(100e-6)
nn.set_state_monitor(state_monitor)
nn.add_neuron_population(num_neurons, neuron)
spike_monitor.reset_spikes()


# Instantiate the graph into the neural network
for pre, post, weight, delay in graph:
    syn = pb.Synapse(pre, post, weight, delay)
    nn.add_synapse(syn)

# Add the stored spike events to the schedule
for t, neuron_id, magnitude in spike_storage:
    nn.schedule_spike_event(t, neuron_id, magnitude)
```

```python
# Running a timed simulation
start_time = time.perf_counter()  # start timer
nn.run(simulation_duration)
end_time = time.perf_counter()    # end timer

print(f"Simulation finished in {end_time - start_time:.6f} seconds")
print(f"Number of events: {len(spike_monitor.spike_list)}")
```

```python
# Compute spike counts per neuron
spike_counts = np.bincount(neuron_ids, minlength=num_neurons)

# Compute firing rates (Hz)
spike_rates = spike_counts / simulation_duration  # spikes per second

# Create bar plot
plt.figure(figsize=(10, 4))
plt.bar(np.arange(num_neurons), spike_rates, color="black")
plt.xlabel("Neuron index")
plt.ylabel("Firing rate (Hz)")
plt.title("Spike Rates per Neuron - Ours")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

## Comparison to Brian 2

```python
from brian2 import *

# Parameters
tau_m = 20*ms
C = 200*pF
v_rest = -70*mV
v_reset = -70*mV
v_thresh = -50*mV
refractory_period = 5*ms

defaultclock.dt = 100*us

eqs = '''
dv/dt = (v_rest - v) / tau_m : volt (unless refractory)
'''
G = NeuronGroup(
    num_neurons,
    eqs,
    threshold='v>v_thresh',
    reset='v=v_reset',
    refractory=refractory_period,
    method='exact'
)
G.v = v_rest

S = Synapses(G, G, 'w : volt', on_pre='v_post += w')

for pre, post, weight, delay in graph:
    S.connect(i=pre, j=post)
    S.w[pre, post] = (weight/C_m)*volt
    S.delay[pre, post] = delay*second

# Convert spike_storage → arrays
indices = np.array([neuron_id for _, neuron_id, _ in spike_storage])
times = np.array([t for t, _, _ in spike_storage]) * second

# Align times to simulation clock resolution
dt = defaultclock.dt
times = np.round(times / dt) * dt

# Remove duplicates (neuron, time) that would cause errors
pairs = set()
clean_indices, clean_times = [], []
for idx, t in zip(indices, times):
    if (idx, float(t)) not in pairs:
        pairs.add((idx, float(t)))
        clean_indices.append(idx)
        clean_times.append(t)

P = SpikeGeneratorGroup(input_neurons, indices=clean_indices, times=clean_times)
SP = Synapses(P, G, on_pre='v_post += 30*mV')  # input strength
SP.connect(i=np.arange(input_neurons), j=np.arange(input_neurons))

spikemon = SpikeMonitor(G)
statemon = StateMonitor(G, 'v', record=True)  # record neuron 0 as example

# Phase 1: Code generation + compilation
start_time = time.perf_counter()
device.build(directory='output', compile=True, run=False)
end_time = time.perf_counter()
print(f"Compile time: {end_time - start_time:.6f} seconds")

# Phase 2: Running a timed simulation
start_time = time.perf_counter()  # start timer
run(simulation_duration*second)
end_time = time.perf_counter()    # end timer
print(f"Simulation finished in {end_time - start_time:.6f} seconds")
print(f"Number of spike events: {spikemon.num_spikes}")
```

```python
# Raster plot comparison
plt.figure(figsize=(10,4))
plt.scatter(spikemon.t/second, spikemon.i, s=5, c='black', label='Brian 2')
plt.scatter(spike_times, neuron_ids, c='orange', alpha=0.8, s=5, marker='x', label='Ours')
plt.xlabel('Time (s)')
plt.ylabel('Neuron index')
plt.legend()
plt.title('Raster Plot of Spike Events')
plt.show()

brian_spike_times = spikemon.t / second  # convert to seconds
brian_neuron_ids = spikemon.i
brian_spike_counts = np.bincount(brian_neuron_ids, minlength=num_neurons)
ours_spike_counts = np.bincount(neuron_ids, minlength=num_neurons)

# --- Compute firing rates (Hz) ---
brian_rates = brian_spike_counts / simulation_duration
ours_rates = ours_spike_counts / simulation_duration

# --- Plot comparison ---
x = np.arange(num_neurons)

plt.figure(figsize=(10, 4))
width = 0.4  # bar width

plt.bar(x - width/2, brian_rates, width=width, color='black', alpha=0.7, label='Brian 2')
plt.bar(x + width/2, ours_rates, width=width, color='orange', alpha=0.7, label='Ours')

plt.xlabel('Neuron index')
plt.ylabel('Firing rate (Hz)')
plt.title('Comparison of Spike Rates per Neuron (Brian 2 vs Ours)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

mae = np.abs(brian_rates - ours_rates).mean()
mask = brian_rates != 0
percentual_errors = np.abs(brian_rates[mask] - ours_rates[mask]) / np.abs(brian_rates[mask])
mean_percent_error = percentual_errors.mean() * 100  # convert to %
print(f"Mean Absolute Error in firing rates: {mae:.2f} Hz")
print(f"Mean Absolute Percentage Error: {mean_percent_error:.2f}%")
```

## Comparison to NEST

```python
import nest
import numpy as np
import matplotlib.pyplot as plt

dt = 0.1
input_weight = 30

nest.ResetKernel()
nest.SetKernelStatus({"resolution": dt, "local_num_threads": num_threads})  # simulation resolution in ms

# ==============================
# Create neurons
# ==============================
neurons = nest.Create("iaf_psc_delta", num_neurons, params={
    "C_m": 200.0,
    "tau_m": 20.0,
    "V_th": -50.0,
    "V_reset": -70.0,
    "E_L": -70.0,
    "t_ref": 5.0,
})

nest.SetStatus(neurons, {"V_m": -70.0})

# ==============================
# Connect neurons according to graph
# ==============================
for pre, post, weight, delay in graph:
    weight_pAms = (weight/C_m)*1000  # Convert weight to pA*ms
    delay_ms = delay*1000  # Convert delay to ms
    nest.Connect(neurons[pre], neurons[post], syn_spec={
        "weight": weight_pAms,
        "delay": delay_ms
    })

# ==============================
# Input spike generator
# ==============================
# Convert spike_storage → NEST format
spike_times_dict = {}
for t, neuron_id, _ in spike_storage:
    spike_times_dict.setdefault(neuron_id, []).append(t*1000) # s -> ms

input_generators = []
for neuron_id, times in spike_times_dict.items():
    sg = nest.Create("spike_generator", params={"spike_times": times})
    nest.Connect(sg, neurons[neuron_id], syn_spec={"weight": input_weight, "delay": dt})
    input_generators.append(sg)

# ==============================
# Recording devices
# ==============================
spikemon = nest.Create("spike_recorder")
nest.Connect(neurons, spikemon)

# Optional: membrane potential recording
statemon = nest.Create("voltmeter")
nest.Connect(statemon, neurons[0])  # record first neuron

# ==============================
# Run simulation
# ==============================
start_time = time.perf_counter()  # start timer
nest.Simulate(simulation_duration*1000)
end_time = time.perf_counter()    # end timer
print(f"Simulation finished in {end_time - start_time:.6f} seconds")

# ==============================
# Extract spike data
# ==============================
events = nest.GetStatus(spikemon, "events")[0]
neuron_ids_nest = events["senders"]
spike_times_nest = events["times"]

# ==============================
# Plot raster
# ==============================
neuron_ids_nest_c = np.array(neuron_ids_nest) - 1  # Convert to 0-based index
plt.figure(figsize=(10,4))
plt.scatter(spike_times_nest/1000, neuron_ids_nest_c, s=5, c='black', label="NEST")
plt.scatter(np.array(spike_times), neuron_ids, c='orange', alpha=0.8, s=5, marker='x', label='Ours')
plt.xlabel('Time (s)')
plt.ylabel('Neuron index')
plt.legend()
plt.title('Raster Plot of Spike Events')
plt.show()
```
